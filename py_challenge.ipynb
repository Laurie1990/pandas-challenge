* Create an overview table that summarizes key metrics about each school, including:

#Need to effectively group by, and then perform aggregations.
Reference:
https://stackoverflow.com/questions/48909110/python-pandas-mean-and-sum-groupby-on-different-columns-at-the-same-time

```python
school_summary = pd.DataFrame(school_data_complete.groupby(['school_name', 'type']).agg({'student_name':'count', 'budget':'sum', 'math_score':'mean', 
                'reading_score':'mean' , 'math_pass_series':'sum', 'reading_pass_series':'sum', 'both_pass_series':'sum'}).reset_index())
school_summary['%_passing_maths'] = (school_summary.math_pass_series/school_summary.student_name)*100
school_summary['%_passing_reading'] = (school_summary.reading_pass_series/school_summary.student_name)*100
school_summary['%_passing_both'] = (school_summary.both_pass_series/school_summary.student_name)*100
school_summary['Total_School_Budget']=school_summary.budget/school_summary.student_name
school_summary['Per_Student_Budget']=school_summary.Total_School_Budget/school_summary.student_name
del school_summary['math_pass_series']
del school_summary['reading_pass_series']
del school_summary['both_pass_series']
del school_summary['budget']
school_summary.rename(columns={'type':'school_type', 'student_name':'number_of_students', 'math_score':'avg_math_score', 'reading_score':'avg_reading_score'}, inplace=True)

school_summary
```

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

#create new series based on conditional logic eg. if year level == 9th grade. etc. Need to specify the data type of new series in order to aggregate
# Reference for casting as new datatype given here: https://stackoverflow.com/questions/15891038/change-data-type-of-columns-in-pandas

```python
#create new conditional series 
school_data_complete.loc[school_data_complete["grade"]=="9th", 'Year_9_math']=school_data_complete["math_score"]
school_data_complete.loc[school_data_complete["grade"]=="10th", 'Year_10_math']=school_data_complete["math_score"]
school_data_complete.loc[school_data_complete["grade"]=="11th", 'Year_11_math']=school_data_complete["math_score"]
school_data_complete.loc[school_data_complete["grade"]=="12th", 'Year_12_math']=school_data_complete["math_score"]
#math_scores_grade = school_data_complete.groupby(['school_name']).agg({'Year_9_math':'mean'})
school_data_complete=school_data_complete.astype({'Year_9_math':float, 'Year_10_math':float, 'Year_11_math': float, 'Year_12_math': float})
math_scores_grade = school_data_complete.groupby(['school_name']).agg({'Year_9_math':'mean','Year_10_math':'mean', 'Year_11_math':'mean', 'Year_12_math':'mean'})
math_scores_grade
```

## Reading Score by Grade
